```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import random
```

```python
imagepath = ['../picture/Agfa_DC-504_0_1.JPG',
             '../picture/Agfa_DC-733s_0_452.JPG',
             '../picture/Agfa_DC-830i_0_996.JPG',
             '../picture/Agfa_Sensor505-x_0_1620.JPG',
             '../picture/Agfa_Sensor530s_0_2033.JPG']
# def _int64_feature(value):
#     return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
def _bytes_featrue(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
count=-1
```

```python
with tf.Session() as sess:
    for model in range(3):
        count=count+1
        image_raw_data = tf.gfile.FastGFile(imagepath[model]).read()
        filename = './Records/test.tfrecords'
        writer = tf.python_io.TFRecordWriter(filename)
        img_data = tf.image.decode_jpeg(image_raw_data)
        img = img_data.eval()
        l = img.shape[0]
        w = img.shape[1]
        for i in range(5):
            crop = tf.image.crop_to_bounding_box(img_data,random.randint(0,l-64),random.randint(0,w-64),64,64)
            example = tf.train.Example(features=tf.train.Features(feature={
                'pixels': _int64_feature(64*64),
                'label': _int64_feature(int(count)),
                'image_raw': _bytes_featrue(crop.eval().tostring()),
            }))
            print example.SerializeToString()
            writer.write(example.SerializeToString())
            #print model,i
        print 'picutre',model,'ready'
            
    writer.close()
```
